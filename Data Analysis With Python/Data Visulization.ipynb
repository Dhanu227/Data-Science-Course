## Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## create a line plot
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Basic Line Plot')
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## customize line plot

plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=8)
```

```python
## Multiple plot

x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(10,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title('Plot 1')

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title('Plot 2')

plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title('Plot 3')

plt.subplot(2,2,4)
plt.plot(y1,y2,color='red')
plt.title('Plot 4')
```

```python
## Bar plot

categories=['A','B','C','D','E']
value=[5,6,7,5,3]

plt.bar(categories,value,color='purple')
plt.xlabel('categories')
plt.ylabel('value')
plt.title('Bar Plot')
```

## Histograms



```python
data=[1,2,1,3,1,4,3,5,3,6,5,6]

plt.hist(data,bins=5,color='orange',edgecolor='black')
```

```python
## create a scatter plot

x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.scatter(x,y,color='blue',marker='x')
```

```python
## pie chart

label=['a','b','c','d']
sizes=[30,50,60,70]
color=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1,0,0,0) ## moves out the 1st slice

## create a pie chart

plt.pie(sizes,explode=explode,labels=label,colors=color,autopct='%1.1f%%')
```

```python
## sales data visulization

import pandas as pd
sales_data=pd.read_csv('sales_data.csv')
sales_data.head()
```

```python
sales_data.info()
```

```python
## plot total sales by product

total_sales_by_product=sales_data.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_product
```

```python
total_sales_by_product.plot(kind='bar',color='teal')
```

```python
## plot sales trend over time

sales_trend=sales_data.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```
