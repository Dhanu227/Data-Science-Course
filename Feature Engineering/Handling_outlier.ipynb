## 5 Number summarya and Box Plot

```python
## Minimum, maximum,median,Q1,Q2,IQR

import numpy as np
```

```python
lst_marks=[45,65,54,55,74,54,56,34,67,75,35,98,56]
minimum,Q1,median,Q3,maximum = np.quantile(lst_marks,[0,0.25,0.50,0.75,1.0])
```

```python
minimum,Q1,median,Q3,maximum
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
lower_fence=Q1-1.5*(IQR)
higher_fence=Q1+1.5*(IQR)
print(lower_fence)
print(higher_fence)
```
